```python
from IPython.display import HTML

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style

import numpy as np
from numpy.random import shuffle, uniform

from sklearn.datasets.samples_generator import make_blobs

import time
```

```python
def create_blobs(centers, random_state):
    np.random.seed(random_state)
    
    offset = uniform(-0.25, 0.25)
    low = 0.4 + offset
    high = 1.6 + offset
    std = uniform(low=low, high=high, size=centers)
    return make_blobs(n_samples=1500, centers=centers, cluster_std=std, shuffle=True, random_state=random_state)
```

```python
np.random.seed(0)

X1, _ = create_blobs(centers=5, random_state=1)
X2, _ = create_blobs(centers=2, random_state=2)
X3, _ = create_blobs(centers=3, random_state=3)

X = np.concatenate((X1, X2, X3))
for _ in range(3):
    shuffle(X)
    
max_x, min_x = max(X[:, 0]) + 3, min(X[:, 0]) - 3
max_y, min_y = max(X[:, 1]) + 3, min(X[:, 1]) - 3
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X1[:, 0], X1[:, 1], 'o')
ax.set_xlim(min_x, max_x)
ax.set_ylim(min_y, max_y)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X2[:, 0], X2[:, 1], 'o')
ax.set_xlim(min_x, max_x)
ax.set_ylim(min_y, max_y)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X3[:, 0], X3[:, 1], 'o')
ax.set_xlim(min_x, max_x)
ax.set_ylim(min_y, max_y)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(X[:, 0], X[:, 1], 'o')
ax.set_xlim(min_x, max_x)
ax.set_ylim(min_y, max_y)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

start, end = 0, 25

def animate(i):
    global start, end
    
    ax.clear()
    ax.plot(X[start:end, 0], X[start:end, 1], 'o')
    ax.set_xlim(min_x, max_x)
    ax.set_ylim(min_y, max_y)
    
    end += 25
    if start + 150 < end:
        start = end - 150
    
anim = animation.FuncAnimation(fig, animate, interval=750)

try:
    HTML(anim.to_html5_video())
except Exception:
    pass
```
